```python
# 필요한 library
import os
from re import A
import subprocess
from natsort import natsorted
import pandas as pd
import keyboard
import signal
import subprocess
import psutil
import math
import shutil
import argparse
from tqdm.notebook import tqdm
from mpl_toolkits.axes_grid1 import host_subplot
import mpl_toolkits.axisartist as AA
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
# 필요한 디렉토리 만들기
# 폴더 있어도 새로 만들기
def refresh_dir(file_path):
    if os.path.exists(file_path):
        os.remove(file_path)
    os.makedirs(file_path)
# 폴더 없으면 
def make_dir(file_path):
    if not os.path.exists(file_path):
        os.makedirs(file_path)
# 컬럼 새로 만들기
def makeNewColumns():
    SIDEIDX = ['non','oa']
    PARTIDX = ['shank','shoe','thigh']
    TYPEIDX = ['ACC','GYRO','MAG']
    AXISIDX = ['X', 'Y', 'Z']
    LEGCOLUMNSLENGTH = 54
    COl_imu_legs = [f'{SIDEIDX[int(i//(LEGCOLUMNSLENGTH/2))]}\
_{PARTIDX[(i//(len(TYPEIDX)*len(AXISIDX)))%len(PARTIDX)]}\
_{TYPEIDX[(i//(len(AXISIDX)))%len(TYPEIDX)]}\
_{AXISIDX[i%len(AXISIDX)]}' for i in range(0,LEGCOLUMNSLENGTH)]
    TRKCOLUMNSLENGTH = 9
    Col_imu_trunk = [f'trunk_{TYPEIDX[(i//(len(AXISIDX)))%len(TYPEIDX)]}_{AXISIDX[i%len(AXISIDX)]}' for i in range(0,TRKCOLUMNSLENGTH)]
    # Columns for forceplate
    FPCOLUMNSLENGTH = 12
    FPTYPEIDX = ['GRF','ANGLE','MONM','MOBWHT']
    Col_FP = [f'{FPTYPEIDX[(i//(len(AXISIDX)))%len(FPTYPEIDX)]}_{AXISIDX[i%len(AXISIDX)]}' for i in range(0,FPCOLUMNSLENGTH)]
    # 최종 column
    newColumns = COl_imu_legs+Col_imu_trunk+Col_FP
    return newColumns

# 필터
def butter_lowpass_filter(data, cutoff, fs, order):
    nyq = 0.5*fs
    normal_cutoff=cutoff/nyq
    b, a = signal.butter(order, normal_cutoff, btype='lowpass', analog=False)
    y = signal.filtfilt(b, a, data)
    return y 

def butter_bandpass_filter(data, cutoff_low, cutoff_high, fs, order):
    nyq = 0.5*fs
    low_cutoff=cutoff_low/nyq
    high_cutoff=cutoff_high/nyq
    b, a = signal.butter(order, [low_cutoff, high_cutoff], btype='bandpass', analog=False)
    y = signal.filtfilt(b, a, data)
    return y 

def filt_onlyIMU(datum):
    newColumns = makeNewColumns()
    datum.columns = newColumns
    # filter 처리한 새로운 dataframe 만들기
    Filt_datum = pd.DataFrame(columns=newColumns)
    for i in range(0,63): # IMU 데이터만 filtering 하기
        raw_Data = datum.iloc[:,i]
        filt_Data = butter_lowpass_filter(raw_Data, 12, SR_DATA, ORDER) # lowpass butter 2nd filter 12Hz Zero-phase shift
        Filt_datum[datum.columns[i]] = filt_Data
        
    for i in range(63,75):
        Filt_datum[datum.columns[i]] = datum.iloc[:,i]
    return Filt_datum
```

```python
######## 필요 변수 정의
SR_DATA = 250 # 현재 해당 폴더내의 모든 데이터는 250HZ로 downsampling 되어 있음
# cutoff frequency
CUTOFF = 12 #이거는 force plate의 sampling rate과 맞춰봄
ORDER = 2
```

```python
###########################
# 한번 했으면 좀 넘어가라!#
###########################
# 파일 경로 설정
dataDir =     r'Z:\PROJECTS\iwalqq\Data\V3D\Output\IMU Deep Learning\Data\allnew_20220325_raw_byDeepak_csv\INC_ByStep\INC_ByZero\Included_checked'
rawDir =      os.path.join(dataDir, r'RAW')
filteredDir = os.path.join(dataDir, r'FILT')

# 절대로 rawDir는 여기서 만들면 안됨
# 이전 단계에서 미리 만들어져있어야 함
make_dir(filteredDir)
dataExt = r".csv"
# 파일목록 가져오기
# 파일 목록 전체 들고오고 필요한 확장자만 고르고 들고온 파일 목록에서 .txt만 남기고 나머지 것들 제외시키기
dataList = natsorted([_ for _ in os.listdir(rawDir) if _.endswith(dataExt)])
# 데이터 저장할 dataframe 만들기
list_data = pd.DataFrame(columns=["patientID", "dateVisit", "speed","numtrial","side","numStep"])
listName  = r'list_dataset_correction.xlsx'
```

```python
# 정리된 데이터 목록화
for file in dataList:
    patientID = file.split("_")[0]
    dateVisit = file.split("_")[1]
    speed     = file.split("_")[2]
    Numtrial  = file.split("_")[3]
    side      = file.split("_")[4]
    NumStep   = file.split("_")[-2]

    df_datum = pd.DataFrame([[patientID, dateVisit, speed, Numtrial, side, NumStep]],columns=["patientID", "dateVisit", "speed","numtrial","side","numStep"])
    list_data = pd.concat([list_data,df_datum],ignore_index=True)
list_data.to_excel(os.path.join(dataDir,listName),index=False,engine = 'openpyxl')
list_data.to_csv("afterPDFCHKforSenddance.csv",index=False)
```

```python
# 정리된 필터 적용해보기
newColumns = makeNewColumns()

for file in tqdm(dataList):
    # column 자르고 이름 깔끔하게 변경하기
    # 데이터가 string이니까 float로 바꾸기
    datum = pd.read_csv(os.path.join(rawDir, file)).iloc[4:,:].reset_index(drop=True).astype(float)
    # IMU만 필터 처리 완료!
    filt_datum = filt_onlyIMU(datum)
    nameforfilt = "F_"+ file
    filt_datum.to_csv(os.path.join(filteredDir,nameforfilt),index=False)
```
